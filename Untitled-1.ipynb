```python
import os, random, json, math, warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt

from ta.trend import SMAIndicator, EMAIndicator, MACD
from ta.momentum import RSIIndicator, StochasticOscillator
from ta.volatility import BollingerBands

from sklearn.preprocessing import StandardScaler
from sklearn.metrics import roc_auc_score, f1_score, precision_recall_fscore_support, accuracy_score, classification_report
from sklearn.model_selection import TimeSeriesSplit

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

SEED = 42
np.random.seed(SEED); random.seed(SEED); tf.random.set_seed(SEED)
```

```python
# Global config 
CFG = {
    "TICKER": "CBA.AX",            # Australian stock (ASX)
    "START": "2010-01-01",
    "END": None,                   # None = today
    "LOOKBACK": 60,                # window length in trading days
    "PRED_HORIZON": 1,             # predict next-day move
    "THRESHOLD": 0.55,             # trade when model prob > threshold
    "TX_COST_BPS": 5,              # round-trip cost in basis points (0.05%)
    "N_SPLITS": 5,                 # walk-forward CV splits
    "BATCH_SIZE": 128,
    "EPOCHS": 30,
    "PATIENCE": 5,
    "LR": 1e-3
}
print(json.dumps(CFG, indent=2))
```

```python
def _normalize_ohlcv(df: pd.DataFrame) -> pd.DataFrame:
    out = df.copy()

    # 1) Flatten potential MultiIndex and clean names
    if isinstance(out.columns, pd.MultiIndex):
        out.columns = ["_".join([str(c) for c in col if c not in (None, "")]).strip().lower()
                       for col in out.columns]
    else:
        out.columns = [str(c).strip().lower() for c in out.columns]

    # 2) Remove exact duplicate columns (keep first)
    out = out.loc[:, ~out.columns.duplicated()].copy()

    # 3) Map common variants to ohlcv canonical names
    rename_map = {
        "adj close": "adj_close",
        "close*": "close",  # handle accidental stars/variants if present
    }
    out = out.rename(columns=rename_map)

    # 4) If label selection returns a DataFrame (duplicate names), squeeze to 1-D
    def _ensure_series(colname):
        obj = out[colname]
        if isinstance(obj, pd.DataFrame):  # duplicate-named columns
            if obj.shape[1] > 1:
                # take the first; optionally warn
                obj = obj.iloc[:, 0]
            else:
                obj = obj.squeeze(axis=1)
        return pd.to_numeric(obj, errors="coerce")

    for c in ["open", "high", "low", "close", "volume"]:
        if c in out.columns:
            out[c] = _ensure_series(c)

    return out
```

```python
def load_asx_history(ticker, start, end=None):
    df = yf.download(ticker, start=start, end=end, auto_adjust=True)
    df = df.rename(columns=str.lower).dropna()
    return df

df = load_asx_history(CFG["TICKER"], CFG["START"], CFG["END"])
df.head(), df.tail(), df.shape
```

```python
import re

def _flatten_cols(cols):
    if isinstance(cols, pd.MultiIndex):
        cols = ["_".join([str(c) for c in col if c not in (None, "")]) for col in cols]
    return [str(c).strip().lower() for c in cols]

def _first_match(cols, base):
    """
    Return the first column that looks like <anything>_<base> OR exactly <base>
    OR contains the base as a full token. e.g., 'close', 'cba.ax_close', 'close_cba.ax'
    """
    exact = [c for c in cols if c == base]
    if exact: return exact[0]
    # token-based contains (split on non-alphanum)
    token_hit = []
    for c in cols:
        tokens = re.split(r"[^a-z0-9]+", c)
        if base in tokens:
            token_hit.append(c)
    if token_hit: return token_hit[0]
    # suffix/prefix with base
    suffix = [c for c in cols if c.endswith(f"_{base}") or c.startswith(f"{base}_")]
    if suffix: return suffix[0]
    # loose contains as last resort
    loose = [c for c in cols if base in c]
    return loose[0] if loose else None

def _normalize_ohlcv(df: pd.DataFrame) -> pd.DataFrame:
    out = df.copy()
    out.columns = _flatten_cols(out.columns)
    out = out.loc[:, ~pd.Index(out.columns).duplicated()].copy()

    # Normalise common variants
    out = out.rename(columns={"adj close": "adj_close"})

    cols = list(out.columns)

    col_open   = _first_match(cols, "open")
    col_high   = _first_match(cols, "high")
    col_low    = _first_match(cols, "low")
    col_close  = _first_match(cols, "close")
    col_volume = _first_match(cols, "volume")
    col_adj    = _first_match(cols, "adj_close")

    # If we don't have 'close' but have 'adj_close', use that
    if col_close is None and col_adj is not None:
        out["close"] = pd.to_numeric(out[col_adj], errors="coerce")
        col_close = "close"
    elif col_close is not None:
        out["close"] = pd.to_numeric(out[col_close], errors="coerce")
    else:
        # Nothing close-like found — raise with helpful info
        raise KeyError(f"No close/adj_close-like column found. Got columns: {cols}")

    # Optional but helpful: set other canonical columns if present
    if col_open is not None:   out["open"]   = pd.to_numeric(out[col_open], errors="coerce")
    if col_high is not None:   out["high"]   = pd.to_numeric(out[col_high], errors="coerce")
    if col_low is not None:    out["low"]    = pd.to_numeric(out[col_low], errors="coerce")
    if col_volume is not None: out["volume"] = pd.to_numeric(out[col_volume], errors="coerce")

    # Drop obvious non-price fields if yfinance added them
    drop_noise = [c for c in out.columns if c in ("dividends", "stock splits", "stock_splits")]
    out = out.drop(columns=drop_noise, errors="ignore")

    # Keep chronological order, drop full-NaN rows
    out = out.sort_index().dropna(how="all")
    return out

def add_technical_indicators(df: pd.DataFrame) -> pd.DataFrame:
    out = _normalize_ohlcv(df)
    close = out["close"]

    # Basic returns
    out["ret_1"] = close.pct_change()
    out["log_ret_1"] = np.log1p(out["ret_1"])

    # Trend
    out["sma_10"] = SMAIndicator(close, window=10).sma_indicator()
    out["sma_20"] = SMAIndicator(close, window=20).sma_indicator()
    out["ema_10"] = EMAIndicator(close, window=10).ema_indicator()
    out["ema_20"] = EMAIndicator(close, window=20).ema_indicator()

    # Momentum
    out["rsi_14"] = RSIIndicator(close, window=14).rsi()
    if all(k in out.columns for k in ("high","low")):
        stoch = StochasticOscillator(high=out["high"], low=out["low"], close=close, window=14, smooth_window=3)
        out["stoch_k"] = stoch.stoch()
        out["stoch_d"] = stoch.stoch_signal()

    # MACD
    macd = MACD(close, window_slow=26, window_fast=12, window_sign=9)
    out["macd"] = macd.macd()
    out["macd_signal"] = macd.macd_signal()
    out["macd_hist"] = macd.macd_diff()

    # Bollinger
    bb = BollingerBands(close, window=20, window_dev=2)
    out["bb_high"] = bb.bollinger_hband()
    out["bb_low"]  = bb.bollinger_lband()
    denom = (out["bb_high"] - out["bb_low"]).replace(0, np.nan)
    out["bb_pct"]  = (close - out["bb_low"]) / denom

    # Volatility & relatives
    out["vol_10"] = out["ret_1"].rolling(10).std()
    out["vol_20"] = out["ret_1"].rolling(20).std()
    out["close_sma10"] = close / out["sma_10"] - 1
    out["close_sma20"] = close / out["sma_20"] - 1

    out = out.dropna().copy()
    return out

# Re-run
df_feat = add_technical_indicators(df)
print(df_feat.columns.tolist()[:20], "...", len(df_feat.columns), "cols")
df_feat.head()
```

```python
## Labeling (classification: next-day direction)
def make_labels(df: pd.DataFrame, horizon=1):
    out = df.copy()
    future_ret = out["close"].pct_change(periods=horizon).shift(-horizon)
    out["y_up"] = (future_ret > 0).astype(int)  # 1 if next-day close goes up
    return out.dropna()

data = make_labels(df_feat, CFG["PRED_HORIZON"])
features = [
    c for c in data.columns
    if c not in ["y_up","open","high","low","close","adj close","volume"]
]
len(features), features[:10]
```

```python
# Build lookback windows (supervised sequences)
def build_sequences(df: pd.DataFrame, feature_cols, target_col, lookback):
    X_seq, y = [], []
    vals = df[feature_cols].values
    y_vals = df[target_col].values
    for i in range(lookback, len(df)):
        X_seq.append(vals[i-lookback:i, :])
        y.append(y_vals[i])
    X_seq = np.array(X_seq, dtype=np.float32)
    y = np.array(y, dtype=np.int32)
    return X_seq, y

raw = data.copy()
X_seq_raw, y_raw = build_sequences(raw, features, "y_up", CFG["LOOKBACK"])
dates = raw.index[CFG["LOOKBACK"]:]  # aligns with X_seq_raw/y_raw
X_seq_raw.shape, y_raw.shape, dates[0], dates[-1]
```

```python
# Time-series scaling (fit on train folds only)
def scale_by_fit(train_2d, test_2d):
    scaler = StandardScaler()
    scaler.fit(train_2d)
    return scaler, scaler.transform(train_2d), scaler.transform(test_2d)

def scale_sequences_train_test(X_train, X_test):
    # flatten to 2D, scale feature-wise, then reshape back
    n_tr, lb, nf = X_train.shape
    n_te = X_test.shape[0]
    scaler, Xtr2d, Xte2d = scale_by_fit(X_train.reshape(n_tr, lb*nf), X_test.reshape(n_te, lb*nf))
    Xtr = Xtr2d.reshape(n_tr, lb, nf).astype(np.float32)
    Xte = Xte2d.reshape(n_te, lb, nf).astype(np.float32)
    return scaler, Xtr, Xte

X_seq_raw.shape
```

```python
def build_cnn_1d(input_shape, lr=1e-3):
    inp = layers.Input(shape=input_shape)
    x = layers.Conv1D(64, kernel_size=5, padding="causal", activation="relu")(inp)
    x = layers.Conv1D(64, kernel_size=5, padding="causal", activation="relu")(x)
    x = layers.MaxPooling1D(pool_size=2)(x)
    x = layers.Conv1D(128, kernel_size=3, padding="causal", activation="relu")(x)
    x = layers.GlobalAveragePooling1D()(x)
    x = layers.Dropout(0.3)(x)
    x = layers.Dense(64, activation="relu")(x)
    out = layers.Dense(1, activation="sigmoid")(x)

    model = keras.Model(inputs=inp, outputs=out)
    model.compile(
        optimizer=keras.optimizers.Adam(learning_rate=lr),
        loss="binary_crossentropy",
        metrics=[keras.metrics.AUC(name="auc"), keras.metrics.BinaryAccuracy(name="acc")]
    )
    return model

input_shape = (CFG["LOOKBACK"], len(features))
model = build_cnn_1d(input_shape, CFG["LR"])
model.summary()
```

```python
tscv = TimeSeriesSplit(n_splits=CFG["N_SPLITS"])
fold_metrics = []
oof_pred = np.zeros_like(y_raw, dtype=np.float32)

for fold, (tr_idx, te_idx) in enumerate(tscv.split(X_seq_raw)):
    X_tr, X_te = X_seq_raw[tr_idx], X_seq_raw[te_idx]
    y_tr, y_te = y_raw[tr_idx], y_raw[te_idx]

    _, X_tr_s, X_te_s = scale_sequences_train_test(X_tr, X_te)

    m = build_cnn_1d(input_shape, CFG["LR"])
    cb = [
        keras.callbacks.EarlyStopping(monitor="val_auc", mode="max",
                                      patience=CFG["PATIENCE"], restore_best_weights=True)
    ]
    hist = m.fit(
        X_tr_s, y_tr,
        validation_split=0.2,
        epochs=CFG["EPOCHS"],
        batch_size=CFG["BATCH_SIZE"],
        verbose=0,
        callbacks=cb
    )

    proba = m.predict(X_te_s, verbose=0).ravel()
    oof_pred[te_idx] = proba
    auc = roc_auc_score(y_te, proba)
    y_hat = (proba >= 0.5).astype(int)
    acc = accuracy_score(y_te, y_hat)
    f1 = f1_score(y_te, y_hat)
    fold_metrics.append({"fold": fold+1, "AUC": auc, "ACC": acc, "F1": f1})
    print(f"Fold {fold+1}: AUC={auc:.3f}, ACC={acc:.3f}, F1={f1:.3f}")

pd.DataFrame(fold_metrics)
```

```python
## OOF performance summary & calibration glance
auc = roc_auc_score(y_raw, oof_pred)
y_hat = (oof_pred >= 0.5).astype(int)
acc = accuracy_score(y_raw, y_hat)
prec, recall, f1, _ = precision_recall_fscore_support(y_raw, y_hat, average="binary")
print(f"OOF AUC={auc:.3f}  ACC={acc:.3f}  F1={f1:.3f}  Precision={prec:.3f}  Recall={recall:.3f}")

print("\nClassification report (OOF):\n", classification_report(y_raw, y_hat, digits=3))

plt.figure(figsize=(6,4))
plt.hist(oof_pred[y_raw==1], bins=30, alpha=0.6, label="Up (y=1)")
plt.hist(oof_pred[y_raw==0], bins=30, alpha=0.6, label="Down (y=0)")
plt.title("OOF predicted probabilities")
plt.legend(); plt.show()
```

```python
## Simple trading backtest (thresholded long-only)
def backtest_threshold(dates, close, proba, threshold=0.55, tx_cost_bps=5, horizon=1):
    # Signal: long when proba >= threshold, flat otherwise
    signal = (proba >= threshold).astype(int)

    # Align to next-day returns
    ret = pd.Series(close).pct_change(periods=horizon).shift(-horizon).values
    # Position applies at t -> earns return at t+1
    strat_ret_gross = signal * ret

    # Transaction costs when signal changes (enter/exit)
    turns = np.abs(np.diff(signal, prepend=0))
    # round-trip cost approximated as cost on change events
    cost = turns * (tx_cost_bps / 1e4)
    strat_ret_net = strat_ret_gross - cost

    # Equity curve
    eq = (1 + strat_ret_net[np.isfinite(strat_ret_net)]).cumprod()
    eq_idx = dates[:len(eq)]

    # Metrics
    ann_factor = 252 / horizon
    mu = np.nanmean(strat_ret_net) * ann_factor
    sigma = np.nanstd(strat_ret_net) * np.sqrt(ann_factor)
    sharpe = mu / (sigma + 1e-12)

    # Max drawdown
    cum = (1 + pd.Series(strat_ret_net)).cumprod()
    peak = cum.cummax()
    dd = (cum/peak - 1).min()

    out = {
        "avg_daily_ret": np.nanmean(strat_ret_net),
        "sharpe": sharpe,
        "max_drawdown": dd,
        "hit_rate": ( (signal==1) & (ret>0) ).sum() / max((signal==1).sum(),1)
    }
    curve = pd.Series(eq, index=eq_idx)
    return out, curve, pd.Series(signal, index=dates)

bt_metrics, equity_curve, signal_series = backtest_threshold(
    dates=dates, close=raw.loc[dates, "close"].values,
    proba=oof_pred, threshold=CFG["THRESHOLD"],
    tx_cost_bps=CFG["TX_COST_BPS"], horizon=CFG["PRED_HORIZON"]
)
bt_metrics, equity_curve.tail()
```

```python
print("Backtest metrics:")
for k,v in bt_metrics.items():
    print(f"  {k}: {v:.4f}")

plt.figure(figsize=(8,4))
equity_curve.plot()
plt.title(f"Equity Curve — {CFG['TICKER']} (threshold={CFG['THRESHOLD']}, cost={CFG['TX_COST_BPS']} bps)")
plt.ylabel("Equity (normalized)"); plt.xlabel("Date")
plt.show()

plt.figure(figsize=(10,1.5))
signal_series.astype(float).plot()
plt.title("Model Long Signal (1=long, 0=flat)")
plt.yticks([0,1]); plt.show()
```

```python
# === Cell 10: Final train on full history & save artifacts (portable) ===
from pathlib import Path

# 1) Refit scaler and model on ALL available data
n = X_seq_raw.shape[0]
scaler_all, X_all_s, _ = scale_sequences_train_test(X_seq_raw, X_seq_raw[-1:])  # scaler fit on all
final_model = build_cnn_1d(input_shape, CFG["LR"])
cb = [keras.callbacks.EarlyStopping(monitor="auc", mode="max",
                                    patience=CFG["PATIENCE"], restore_best_weights=True)]
final_model.fit(
    X_all_s, y_raw,
    epochs=CFG["EPOCHS"],
    batch_size=CFG["BATCH_SIZE"],
    verbose=0,
    callbacks=cb
)

# 2) Choose a safe writable directory (here: /tmp)
save_dir = Path("/tmp/artifacts")
save_dir.mkdir(parents=True, exist_ok=True)

# Filenames (use .keras per Keras 3+ recommendation)
model_path = save_dir / "model_tf_cnn1d.keras"
config_path = save_dir / "config.json"
features_path = save_dir / "features.csv"
scaler_path = save_dir / "scaler.json"

# 3) Save model, config, features, and scaler state
final_model.save(str(model_path))  # native Keras format
with open(config_path, "w") as f:
    json.dump(CFG, f, indent=2)

pd.Series(features).to_csv(features_path, index=False)

scaler_state = {
    "mean": scaler_all.mean_.tolist(),
    "scale": scaler_all.scale_.tolist(),
    "n_features": int(np.prod(X_seq_raw.shape[1:]))
}
with open(scaler_path, "w") as f:
    json.dump(scaler_state, f, indent=2)

print("✅ Saved artifacts:")
print("  Model   :", model_path)
print("  Config  :", config_path)
print("  Features:", features_path)
print("  Scaler  :", scaler_path)
```

```python
# === Cell 11: Inference + Mini Backtest (loads from /tmp/artifacts) ===
from pathlib import Path
import json, numpy as np, pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras

# ------------------------------------------------------------
# Paths & config
# ------------------------------------------------------------
save_dir   = Path("/tmp/artifacts")
model_path = save_dir / "model_tf_cnn1d.keras"
scaler_path= save_dir / "scaler.json"

THRESHOLD   = CFG["THRESHOLD"]
TX_COST_BPS = CFG["TX_COST_BPS"]
LOOKBACK    = CFG["LOOKBACK"]
HORIZON     = CFG["PRED_HORIZON"]

# ------------------------------------------------------------
# Helpers
# ------------------------------------------------------------
def prepare_latest_window(df_full, features, lookback):
    """Extract the most recent lookback-window of features for prediction."""
    arr = df_full[features].tail(lookback).values.astype(np.float32)
    if arr.shape[0] < lookback:
        raise ValueError(f"Not enough rows to form a window of length {lookback}.")
    return arr[None, :, :]  # (1, lookback, n_features)

def apply_saved_scaler(X_seq, scaler_state):
    """Apply previously fitted StandardScaler parameters (mean, scale) to X_seq."""
    flat = X_seq.reshape(X_seq.shape[0], -1)  # (n_samples, lookback * n_features)
    mean = np.array(scaler_state["mean"])
    scale = np.array(scaler_state["scale"])
    flat_s = (flat - mean) / (scale + 1e-12)
    return flat_s.reshape(X_seq.shape)

def build_sequences_matrix(df: pd.DataFrame, feature_cols, lookback):
    """Create overlapping windows across the whole dataset for batch inference."""
    vals = df[feature_cols].values.astype(np.float32)
    Xseq = []
    idx  = []
    for i in range(lookback, len(df)):
        Xseq.append(vals[i-lookback:i, :])
        idx.append(df.index[i])
    return np.array(Xseq, dtype=np.float32), pd.Index(idx)

def backtest_from_proba(dates, close, proba, threshold=0.55, tx_cost_bps=5, horizon=1):
    """Long-only threshold strategy backtest."""
    signal = (proba >= threshold).astype(int)

    # Next-period returns (aligned to realization at t+1)
    ret = pd.Series(close).pct_change(periods=horizon).shift(-horizon).values

    # Strategy gross return (position at t earns return at t+1)
    strat_ret_gross = signal * ret

    # Transaction costs on signal changes (enter/exit)
    turns = np.abs(np.diff(signal, prepend=0))
    cost = turns * (tx_cost_bps / 1e4)
    strat_ret_net = strat_ret_gross - cost

    # Equity curve (drop trailing NaNs from horizon shift)
    good = np.isfinite(strat_ret_net)
    eq = (1 + pd.Series(strat_ret_net[good], index=dates[:good.sum()])).cumprod()

    # Metrics
    ann_factor = 252 / horizon
    mu = np.nanmean(strat_ret_net) * ann_factor
    sigma = np.nanstd(strat_ret_net) * np.sqrt(ann_factor)
    sharpe = mu / (sigma + 1e-12)
    cum = (1 + pd.Series(strat_ret_net)).cumprod()
    peak = cum.cummax()
    max_dd = (cum/peak - 1).min()

    hit_rate = ( ((signal==1) & (ret>0)).sum() / max((signal==1).sum(), 1) )

    metrics = {
        "avg_daily_ret": float(np.nanmean(strat_ret_net)),
        "ann_return": float(mu),
        "ann_vol": float(sigma),
        "sharpe": float(sharpe),
        "max_drawdown": float(max_dd),
        "hit_rate_when_long": float(hit_rate),
        "threshold": float(threshold),
        "tx_cost_bps": float(tx_cost_bps)
    }
    return metrics, eq, pd.Series(signal, index=dates)

# ------------------------------------------------------------
# Load artifacts
# ------------------------------------------------------------
with open(scaler_path) as f:
    scaler_state = json.load(f)
loaded = keras.models.load_model(str(model_path))

# ------------------------------------------------------------
# Latest prediction
# ------------------------------------------------------------
latest_X = prepare_latest_window(data, features, LOOKBACK)
latest_X_s = apply_saved_scaler(latest_X, scaler_state)
p_up = float(loaded.predict(latest_X_s, verbose=0).ravel()[0])
signal_now = (p_up >= THRESHOLD)
print(f"Predicted P(up) for next day: {p_up:.3f}  -> Signal={signal_now}")

# ------------------------------------------------------------
# Mini backtest across the full history
# (Uses model fitted on all data → optimistic; good for monitoring, not evaluation.)
# ------------------------------------------------------------
X_all, idx_all = build_sequences_matrix(data, features, LOOKBACK)
X_all_s = apply_saved_scaler(X_all, scaler_state)
proba_all = loaded.predict(X_all_s, verbose=0).ravel()

# Align to dates and prices
dates_seq = idx_all                 # dates for each proba window end
close_seq = data.loc[dates_seq, "close"].values

metrics, equity_curve, signal_series = backtest_from_proba(
    dates=dates_seq,
    close=close_seq,
    proba=proba_all,
    threshold=THRESHOLD,
    tx_cost_bps=TX_COST_BPS,
    horizon=HORIZON
)

print("\nMini backtest metrics (in-sample monitor; not leakage-safe):")
for k,v in metrics.items():
    print(f"  {k:>18}: {v:.4f}" if isinstance(v, float) else f"  {k:>18}: {v}")

# ------------------------------------------------------------
# Plots (last ~180 trading days for readability)
# ------------------------------------------------------------
tail_n = min(180, equity_curve.shape[0])

plt.figure(figsize=(9,4))
equity_curve.iloc[-tail_n:].plot()
plt.title(f"Equity Curve (last {tail_n} days) — {CFG['TICKER']}")
plt.ylabel("Equity (normalized)"); plt.xlabel("Date")
plt.show()

plt.figure(figsize=(9,1.6))
signal_series.iloc[-tail_n:].astype(float).plot()
plt.title("Model Long Signal (1=long, 0=flat) — recent window")
plt.yticks([0,1]); plt.xlabel("Date")
plt.show()

# ------------------------------------------------------------
# Display today’s signal context
# ------------------------------------------------------------
print("\nMost recent signal context:")
print(f"  Date: {data.index[-1].date()}")
print(f"  Close: {data['close'].iloc[-1]:.4f}")
print(f"  P(up): {p_up:.3f}  | Threshold: {THRESHOLD}  | Go long now? {signal_now}")
```
